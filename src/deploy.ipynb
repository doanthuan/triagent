```python
import vertexai
from triagent.agent import root_agent
from vertexai.preview.reasoning_engines import AdkApp
from vertexai import agent_engines


PROJECT_ID = "doanthuan-blog"
LOCATION = "us-central1"
STAGING_BUCKET = "gs://med-agents"

vertexai.init(
    project=PROJECT_ID,
    location=LOCATION,
    staging_bucket=STAGING_BUCKET,
)
```

```python
app = AdkApp(
    agent=root_agent,
    enable_tracing=True,
)
```

```python
session = app.create_session(user_id="u_12345")
session
```

```python
app.list_sessions(user_id="u_12345")
session = app.get_session(user_id="u_12345", session_id="071641f9-8d7a-4a88-9bce-3ce98e660318")
```

```python
for event in app.stream_query(
    user_id="u_12345",
    session_id=session.id,
    message="Verify this claim: Bariatric surgery increases rates of postmenopausal breast cancer. Yes no No",
):
    print(event)
```
